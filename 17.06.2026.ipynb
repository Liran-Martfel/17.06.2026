```python
#get last 4 digit from CC number
df = pd.read_csv ('tips.csv')
def last_four(num):
    return str(num) [-4:]
df ['4 last digit'] = df['CC Number'].apply(last_four)
df
```

```python
#create column first name
#use column name + apply
df = pd.read_csv ('tips.csv')
def first_name(Payer_Name):
    first = str(Payer_Name).split()
    return first [0]
df['first name'] = df['Payer Name'].apply(first_name)
df
```

```python
#create new column - 'table_type'
#if size > 2 -- 'large else if 2 -- regular else -- single
#if smoker -- smokes, else -- healthy
df = pd.read_csv ('tips.csv')
def table_hour (table,smoke):
    if table > 2:
        table = 'Large'
    elif table == 2:
        table =  'regular'
    elif table > 2:
        table = 'single'
    if smoke == 'no':
        smoke = 'healthy'
    else:
        smoke = 'smokes'
    return f"{table},{smoke}"
df ['table_type'] = np.vectorize(table_hour) (df['size'],df['smoker'])
df

    
```
