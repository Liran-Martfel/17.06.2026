```python
#2 - Add a column price_per_person that holds the bill divided by the table size, rounded to 2 decimal places.
df['price_per_person'] = (df['total_bill'] / df['table_size']).round(decimals = 2)
df
```

```python
#3 - Filter the DataFrame to show only orders where price_per_person is greater than 12
df = df[df['price_per_person'] > 12]
df
```

```python
#4 - Use .apply() to create a column table_type: 'Large' if table_size >= 4, otherwise 'Small'
def large(table_size):
    if table_size >= 4:
        return 'Large'
    else:
        return 'Small'
df['table_type'] = df['table_size'].apply(large)
df
```

```python
#5 - Set customer as the index. Then use .loc[] to select the rows from Alice to Carol
df = df.set_index('customer')
df.loc['Alice' : 'Carol']
```

```python
#6 - Use pd.concat with axis=0 to create a new DataFrame containing the first 3 rows and the last 3 rows of the original DataFrame. Reset the index
df1 = pd.concat([df.head(3),df.tail(3)], axis = 0)
df1 = df1.reset_index(drop=True)
df1
```

```python
#7 - Use np.vectorize to create a column value_label: 'High Value' if total_bill > 50 and tip > 10%, otherwise 'Standard'
def check_value(total_bill, tip):
    if total_bill > 50 and tip > (total_bill * 0.1):
        return 'High Value'
    else:
        return 'Standard'

check = np.vectorize(check_value)
df['value_label'] = check(df['total_bill'], df['tip'])
df
```
